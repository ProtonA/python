```python
import random
    
def get_barrel(): #получение случайного бочонка из набора
    a = random.choice(barrels)
    barrels.remove(a)
    return (a)

def get_blank(): #получение случайной карточки
    a = random.sample(range(1, 91), 15) #получаем 15 случайных значений в промежутке 1-90
    a1 = sorted(a[:5]) #делим полученную строку на 3 строки
    a2 = sorted(a[5:10])
    a3 = sorted(a[10:15])
    spaces = sorted(random.sample(range(0, 9), 5)) #случайная растановка пробелов в строке карточки
    j = 0 
    b1 = [' ']*9 #создаем пустую строку
    for i in spaces: #заполняем случайные позиции в пустой строке случайными числами
        b1[i] = a1[j] 
        j += 1        
    spaces = sorted(random.sample(range(0, 9), 5)) 
    j = 0
    b2 = [' ']*9
    for i in spaces:
        b2[i] = a2[j]
        j += 1        
    spaces = sorted(random.sample(range(0, 9), 5))
    j = 0
    b3 = [' ']*9
    for i in spaces:
        b3[i] = a3[j]
        j += 1
    b = b1 + b2 + b3
    return (b)

barrels = [] #создаем набор из 90 боченков
for i in range (1, 91):
    barrels.append(i)

comp_blank = get_blank() #получаем карточку компьютера
player_blank = get_blank() #получаем карточку игрока
new_barrel = get_barrel() #вытаскиваем боченок

print('\n', ' --карточка компьютера:--', '\n', comp_blank[:9], '\n', comp_blank[9:18], '\n', comp_blank[18:27]) #показ карточки
print('\n', ' -----ваша карточка:-----', '\n', player_blank[:9], '\n', player_blank[9:18], '\n', player_blank[18:27])
print ('\n выпал бочонок номер: ', new_barrel, ' осталось бочонков: ', len(barrels))

if new_barrel in comp_blank:
    i = comp_blank.index(new_barrel)
    comp_blank[i] = '-'


while True: 
    
    answer = str(input('Зачеркнуть цифру? (y/n) '))
    
    if (answer == 'y') and (new_barrel not in player_blank):
        print('к сожалению вы проиграли')
        break
        
    if (answer == 'n') and (new_barrel in player_blank):
        print('к сожалению вы проиграли')
        break
        
    if (answer == 'y') and (new_barrel in player_blank):
        i = player_blank.index(new_barrel)
        player_blank[i] = '-'
        new_barrel = get_barrel()
        print('\n', ' --карточка компьютера:--', '\n', comp_blank[:9], '\n', comp_blank[9:18], '\n', comp_blank[18:27])
        print('\n', ' -----ваша карточка:-----', '\n', player_blank[:9], '\n', player_blank[9:18], '\n', player_blank[18:27])
        print('\n выпал бочонок номер: ', new_barrel, ' осталось бочонков: ', len(barrels))
        
    if (answer == 'n') and (new_barrel not in player_blank):
        new_barrel = get_barrel()     
        print('\n', ' --карточка компьютера:--', '\n', comp_blank[:9], '\n', comp_blank[9:18], '\n', comp_blank[18:27])
        print('\n', ' -----ваша карточка:-----', '\n', player_blank[:9], '\n', player_blank[9:18], '\n', player_blank[18:27])
        print('\n выпал бочонок номер: ', new_barrel, ' осталось бочонков: ', len(barrels))
        
    if new_barrel in comp_blank:
        i = comp_blank.index(new_barrel)
        comp_blank[i] = '-'  
        
# проверка выигыша   

    if (player_blank.count('-') == 15) and (comp_blank.count('-') == 15):
        print ('\n !!!НИЧЬЯ!!!')
        break
                
    if player_blank.count('-') == 15:
        print ('\n !!!ВЫ ВЫИГРАЛИ!!!')
        break
        
    if comp_blank.count('-') == 15:
        print ('\n !!!КОМПЬЮТЕР ВЫИГРАЛ!!!')
        break
            
print ('\n несыгравшие бочонки: ', barrels)
```
